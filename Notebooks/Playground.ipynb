```python
import sys
import os

sys.path.append(os.path.abspath(".."))
os.chdir(os.path.abspath(".."))

from fastembed import TextEmbedding
import numpy as np
import faiss
```

```python
embedder = TextEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

```python
def embed(texts):
    return np.array(list(embedder.embed(texts)))

def embed_career(candidate):
    texts = [
        f"{j.company} {j.title} {j.description}"
        for j in candidate.career_history
    ]
    vecs = embed(texts)
    return vecs.mean(axis=0)

def embed_skills(candidate):
    return embed([" ".join([s.name for s in candidate.skills])])[0]

def embed_profile(candidate):
    p = candidate.profile
    text = f"{p.headline} {p.summary} {p.current_title} {p.current_industry}"
    return embed([text])[0]

def build_candidate_vector(candidate):
    c = embed_career(candidate)
    s = embed_skills(candidate)
    p = embed_profile(candidate)

    return np.concatenate([c, s, p])
```

```python
import msgspec
from candidate import Candidate
```

```python
with open("sample/candidates.jsonl", "rb") as fp:
    candidate_list = [msgspec.json.decode(line, type=Candidate) for line in fp]
```

```python
dim = len(build_candidate_vector(candidate_list[0]))
print(f"{dim = }")

index = faiss.IndexFlatL2(dim)

vectors = []

for i, c in enumerate(candidate_list, 1):
    print(f"Building vector for [{i:06}/{len(candidate_list)}]", end="\r")
    v = build_candidate_vector(c)
    vectors.append(v)
print("Vectors built")

vectors = np.array(vectors).astype("float32")

index.add(vectors)
```
